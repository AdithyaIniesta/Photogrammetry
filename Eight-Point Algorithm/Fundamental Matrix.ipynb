```python
import numpy as np
import os
import math
import cv2 
import matplotlib.pyplot as plt
import copy
from random import seed
from random import random
seed(1)
```

```python
class findFundamentalMat():
    
    def __init__(self, image1, image2):
        
        self.image1 = image1
        self.image2 = image2
        self.F = self.computeFMatrix()
    
    @property
    def getFMatrix(self):
        #return Fundamental Matrix
        return self.F
    
    def normalize(self, image):
        
        x , y = image[:,0], image[:,1]
        x_mean, y_mean = np.mean(x), np.mean(y)
        n = len(x)
        rms = np.sum(np.sqrt((x - x_mean)**2 + (y - y_mean)**2))
        rms = rms / n
        
        s = np.sqrt(2) / rms
        t_x = -s * x_mean
        t_y = -s * y_mean
        
        # similarity transform 
        T = np.array([[s, 0, t_x],
                    [0, s, t_y], 
                    [0, 0, 1]])
        
        n_image = list()
        for [i, j] in image:
            n_image.append(T.dot(np.array([i,j,1])))
        
        n_image = np.array(n_image)
        return T, n_image

    def computeFMatrix(self):
        
        #normalize image points
        T1, image1 = self.normalize(self.image1)
        T2, image2 = self.normalize(self.image2)
        
        # Af = 0, compute A
        A = list()
        for i,j in zip(image1, image2):
            x1 , y1 = i[0] , i[1]
            x2 , y2 = j[0] , j[1]
            # A = [x1 * x2, y1 * x2, x2, x1 * y2, y1 * y2, y2, x1, y1, 1 ]
            A.append([ x1 * x2, y1 * x2, x2, x1 * y2, y1 * y2, y2, x1, y1, 1 ])
        
        #compute f by taking SVD of A
        A = np.array(A)
        _, _ , vh = np.linalg.svd(A)
        f = vh.T #last column of v gives f
        F = f[:,-1]
        F = np.reshape(F, (3,3))
        F = F / F[-1][-1]
        
        #replace F by F'
        u , s, vh = np.linalg.svd(F)
        assert(s[0] >= s[1] >= s[2])
        s[2] = 0.0
        S = np.zeros((len(s), len(s)))
        np.fill_diagonal(S, s)
        F_ = u.dot(S.dot(vh))
        F_ = F_ / F_[-1,-1]
        
        #Denormalization F = T2^T * F' * T1
        F = T2.T.dot(F_.dot(T1))
        F = F / F[-1,-1]
        
        return F            
```

```python
def drawCircles(img, center_coordinates, radius = 20, thickness = -2, shift_x = 0, shift_y = 0):
    
    no_of_points = len(center_coordinates)
    temp_img = copy.deepcopy(img)
    
    for i in range(no_of_points):
        [px, py] = center_coordinates[i]
        color = tuple(np.random.randint(0,255,3).tolist())
        temp_img = cv2.circle(temp_img, (int(px + shift_x), int(py + shift_y)), radius, color, thickness)
    
    return temp_img
```

```python
def drawLines(img, start_points, end_points, shift, thickness = 2):
    assert(start_points.shape == end_points.shape)
    
    temp_img = copy.deepcopy(img)
    
    for [sx,sy], [ex, ey] in zip(start_points, end_points):
        color = tuple(np.random.randint(0,255,3).tolist())
        temp_img = cv2.arrowedLine(temp_img,(int(sx), int(sy)), (int(ex + shift), int(ey)),color, thickness)
    
    return temp_img
```

```python
def drawEpilines(img1,lines,pts):

    r,c = img1.shape[:2]

    for r,pt1 in zip(lines,pts):
        pt1 = (int(pt1[0]), int(pt1[1]))
        color = tuple(np.random.randint(0,255,3).tolist())
        x0,y0 = map(int, [0, -r[2]/r[1] ])
        x1,y1 = map(int, [c, -(r[2]+r[0]*c)/r[1] ])
        img1 = cv2.line(img1, (x0,y0), (x1,y1), color,1)
        img1 = cv2.circle(img1, pt1,5,color,-1)
    return img1
```
